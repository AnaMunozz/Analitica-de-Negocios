# **Reto Quincenal: Árbol de Decisión**
Se desarolla un modelo de árbol de decisión que permita pronosticar si una persona tendrá una enfermedad cardiaca, a partir de un conjunto de variables clinicas regstradas en la base de datos. Se busca identificar parones en las características fisiologicas de los individuos, para lo que se utilizaran las siguientes variables:


* **Edad:** La edad que presenta cada persona al momento del registro
*   **Sistótica:** Mida la presión arterial sintólitica, es decir, cuando el corazón se contrae y bombea sangre hacia el resto del cuerpo. Se mide en milimitros de mercurio (mmHg)
*   **Diastólica:** Mide la presión arterial diastólica, es decir, cuando el corazón se encuentra en fase de reposo entre latidos. También se mide en mmHg.
*   **IMC:** El Índice de Masa Corporal es una medida que relaciona el peso y la estatura de una persona, permitiendo clasificar en a las personas en bajo peso, normal, sobrepeso u obesidad.

####**0. Carga de Librerías**
Se procede con la carga de las librerías para el correcto desarrollo del modelo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
```

###**1. Cargar base de datos**
Se carga los datos de trabajo de la base "Base Datos Heart"

```python
nxl = '/content/sample_data/2. HeartDisease.xlsx'
XDB = pd.read_excel(nxl, sheet_name = 0)

XD = XDB [['Age',	'Systolic',	'Diastolic',	'BMI']]  #Variables de trabajo
yd = XDB[['Disease']]                                # Variable de decisión
```

###**2. Modelo de Árbol**
Se procede con la implementación del modelo de árbol

```python
mar = DecisionTreeClassifier(criterion = "gini", max_depth = 4)
mar.fit(XD, yd)  # El modelo busca relación entrada-salida

ydp = mar.predict(XD) # Esto es lo que el modelo pronóstica

#MATRIZ DE CONFUSIÓN
cm = confusion_matrix(yd, ydp)
display(cm)
VN = cm[0,0]; FP = cm[0,1]; FN = cm[1,0]; VP = cm[1,1]

# METRICAS DE DESEMPEÑO
#1. Exactitud: Funcionamiento general del modelo
Ex = (VP+VN)/len(XD)
print("Exactitud: ", Ex)

#2. Tasa Error: %Fallos del modelo
TEr = (FP+FN)/len(XD)
print("Tasa Error: ", TEr)

#3. Sensibilidad: Comportamiento frente a los positivos
Se = VP/(VP+FN)
print("Sensibilidad: ", Se)

#4. Especificidad: Comportamiento frente a los negativos
Es = VN/(VN+FP)
print("Especificidad: ", Es) ##Se preniegan más de los que se preaprueban

#5. Precisión: Versión alternativa del comportamiento frente a los positivos
Pr = VP/(VP+FP)
print("Precisión: ", Pr)

#6. Predicción Negativa: Comportamiento frente a los créditos PreNegados
PrN = VN/(VN+FN)
print("Predicción Negativa: ", PrN)
```

###**3. Despliegue del Árbol de Decisión**

```python
from sklearn.tree import export_graphviz  # Exporta datos a un gráfico
from pydotplus import graph_from_dot_data # Es un graficador

vs = ["Edad",	"Sistólica",	"Diastólica",	"IMC"]      # Títulos del árbol
dot_data = export_graphviz(mar, feature_names = vs)   # Exportar de números a gráficos en PDF
graph = graph_from_dot_data(dot_data)                 # Hacemos el gráfico
graph.write_png("Arbol.png")                          # Visualización del gráfico como png en la carpeta Conetent
```
