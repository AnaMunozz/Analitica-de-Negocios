#**Caso de Estudio 01**

Desarrollar y analizar un modelo Naive Bayes para la detección de riesgo de lavado de activos (SARLAF), utilizando las información de su comportamiento financiero y los cambios en su patrimonio.
En este sentido, las variables a tener en cuenta son:


* **ingresos_usd:** Ingresos percibidos por el cliente como resultado de sus actividades económicas.

* **gastos_usd:** Total de gastos registrados por el cliente en un período determinado.

* **monto_transado_tarjetas_usd:** Valor total de las transacciones realizadas mediante tarjetas de crédito.

* **porcentaje_crecimiento_patrimonio:** Variación porcentual del patrimonio del cliente, utilizada como indicador de cambios inusuales en su situación financiera.




###0. Carga de librerías
Se procede con la carga de librerías de trabajo

```python
import numpy as np
import pandas as pd

#Librerías específicas
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

###1. Carga de datos
Se cargan los datos de trabajo de la base *2. LavadoActivos*

```python
nxl = '/content/2. LavadoActivos.xlsx'
XDB = pd.read_excel(nxl)

#Seleccionamos variables de trabajo
XD = XDB [['ingresos_usd', 'gastos_usd', 'monto_transado_tarjetas_usd', 'porcentaje_crecimiento_patrimonio']] # Variables de trabajo
XD.head()


yd = XDB[['lavado_activos']] # Variable de decisión
yd.head()
```

###2. Implementación Modelo Naive Bayes


```python
mnb = GaussianNB()
mnb.fit(XD, yd) # Ajustar variables Entrada-Salida

#Mostar las medias de las variables
u = mnb.theta_
sigma = mnb.var_
print('ingresos_usd', 'gastos_usd', 'monto_transado_tarjetas_usd', 'porcentaje_crecimiento_patrimonio')
print(u)
print("\nLas desviaciones son: ")
print(sigma)
```

# **Análisis de Resultados**
De los resultados arrojados por el modelo, podemos observar que los individuos con riesgo de lavado de activos (*RiesgoLavado*) presentan, en promedio, **ingresos** considerablemente superiores a los del grupo sin riesgo (*NoRiesgo*), lo que muestra la introducción de montos elevados al sistema financiero por pate de estos clientes.

Así mismo, los **gastos** también son mucho mayores en el grupo *RiesgoLavado*, lo que sugiere un nivel de consumo elevado acorde con los mayores ingresos declarados. Además, el el *monto transado a través de tarjetas de crédito* es mayor en los individuos clasificados como *RiesgoLavado*, lo que muestra su alto uso de productos financieros.

Por otro lado, *ambas categorías* presentan **tasas de creimiento del patrimonio** similares en promedio, aunque los *RiesgoLavado* presentan más desviación estandar, siendo un indicador de que, aunque el crecimiento promedio pueda parecer similar, existen casos extremos de incrementos en el patrimonio en el caso de los clasificados como *RiesgoLavado*.


### 3. Evaluación del modelo | Matriz de Confusión
Se procede con la evaluación del modelo. Para este tipo de modelos, se utiliza la matriz de confusión en su evaluación

```python
ydp = mnb.predict(XD)
cm = confusion_matrix(yd, ydp)
print(cm)

#Métricas de la matriz de confusión
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1]; TDatos=len(XDB)

#1. Exactitud: Funcionamiento general del modelo
Ex = (VP+VN)/TDatos
print("Exactitud: ", Ex)

#2. Tasa Error: %Fallos del modelo
TEr = (FP+FN)/TDatos
print("Tasa Error: ", TEr)

#3. Sensibilidad: Comportamiento frente a los positivos
Se = VP/(VP+FN)
print("Sensibilidad: ", Se)

#4. Especificidad: Comportamiento frente a los negativos
Es = VN/(VN+FP)
print("Especificidad: ", Es) ##Se preniegan más de los que se preaprueban

#5. Precisión: Versión alternativa del comportamiento frente a los positivos
Pr = VP/(VP+FP)
print("Precisión: ", Pr) ##Con esto, el modelo ya funciona un 80% aprobando créditos

#6. Predicción Negativa: Comportamiento frente a los créditos PreNegados
PrN = VN/(VN+FN)
print("Predicción Negativa: ", PrN)
```
